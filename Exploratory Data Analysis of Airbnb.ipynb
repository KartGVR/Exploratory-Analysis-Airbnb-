# Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading and Statistical Analysis of Dataset

```python
# Loading the dataset
df = pd.read_csv("Airbnb_NYC_2019.csv")
df.head()
```

```python
# Number of rows and columns in the dataset
df.shape
```

```python
# Getting information regarding the dataframe
df.info()
```

```python
# Performing statistical operations on the dataset
df.describe()
```

```python
# Different types of data types 
df.dtypes
```

# Cleaning of Dataset

```python
# Finding the missing values in the dataset
df.isnull().sum()
```

```python
# In the given dataset the host_name may be same but the host_id is unique for each host. Hence host_name is removed.
# Last review is removed because it's of no use and has missing values
# host_id is more useful than Id to identify hosts hence it can be removed

df.drop(['id','host_name','last_review'], axis=1, inplace=True)
df.head()
```

# Analysis of Neighbourhood Groups

```python
# Finding unique values of the neighbourhood groups
df.neighbourhood_group.unique()
```

```python
# Plotting a Pie Chart of frequency of Neighbourhood Groups
plt.title("Neighbourhood Group")
x = df.neighbourhood_group.value_counts()
plt.pie(x, labels=x.index, autopct='%1.0f%%')
```

# Analysis of Room Types


```python
# Finding different types of room types used
df.room_type.unique()
```

```python
# Plotting a Pie Chart of frequency of Room Types that are used
plt.title("Room Types")
x = df.room_type.value_counts()
plt.pie(x, labels=x.index, autopct='%1.0f%%')
plt.legend(loc="lower left")
```

# Analysis of Hosts

```python
# The number of times a host appers in present in the column calculated_host_listings_count
# Finding top 10 hosts
# Allthe top 10 hosts are same so the duplicates need to be removed
df.nlargest(10,['calculated_host_listings_count'])
```

```python
# Removing duplication in top 10 hosts
top_10_hosts=df.host_id.value_counts().head(10)
top_10_hosts
```

```python
top_10_hosts=pd.DataFrame(top_10_hosts)
top_10_hosts.reset_index(inplace=True)
top_10_hosts.rename(columns={'index':'Host_ID', 'host_id':'P_Count'}, inplace=True)
top_10_hosts
```

```python
bar_graph=sns.barplot(x="Host_ID", y="P_Count", data=top_10_hosts,
                 palette='Blues_d')
bar_graph.set_title('Top 10 Hosts in terms of Listings')
bar_graph.set_ylabel('Listings')
bar_graph.set_xlabel('Host ID')
bar_graph.set_xticklabels(bar_graph.get_xticklabels(), rotation=45)
```
